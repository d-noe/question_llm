```python
async def collect_answers(
    character_id:str,
    prompts:list[str],
)->list:

    client = Client()
    await client.authenticate(TOKEN)

    answers = []
    for p in tqdm(prompts):
        # instantiate new chat to avoid history
        chat, greeting_message = await client.chat.create_chat(character_id)
        # send message and store answer
        answer = await client.chat.send_message(character_id, chat.chat_id, p)
        answers += [answer.get_primary_candidate().text]

    await client.close_session()

    return answers
```
